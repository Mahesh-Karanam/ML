```python
#FIND S ALGORITHM
class FindS :
  def __init__(self,att):
    self.hypothesis=[0]*att;

  def filter_positive(self,data):
    res=[]
    for i in range(len(data)) :
      if data[i][-1]=='yes':
        res.append(data[i][:-1])
    return res

  def fit(self,positive_examples):
    for example in positive_examples :
      for i, attribute in enumerate(example):
        if self.hypothesis[i]==0:
          self.hypothesis[i]=attribute
        elif self.hypothesis[i]!=attribute :
          self.hypothesis[i]='?'

  def predict(self):
    return self.hypothesis

data=[["sunny","warm","normal","strong","warm","same","yes"],
      ["sunny","warm","high","strong","warm","same","yes"],
      ["rainy","cold","high","strong","warm","change","no"],
      ["sunny","warm","high","strong","cool","change","yes"]]
model=FindS(6)
positive_data=model.filter_positive(data)
model.fit(positive_data)
model.predict()
```

```python
#3. Write a program to implement the naïve Bayesian classifier for a sample training data set stored as a .CSV file.
#Compute the accuracy of the classifier, considering few test data sets.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score

# Load the dataset
data = pd.read_csv("text_classification.csv")

# Split data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.8, random_state=42)

# Prepare data for training
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train_data['text'])
y_train = train_data['review']

# Train the Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train, y_train)

# Prepare test data
X_test = vectorizer.transform(test_data['text'])
y_test = test_data['review']

# Predict on the test data
y_pred = classifier.predict(X_test)

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#7...........back propagation algo
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        self.weights_input_hidden = np.random.randn(input_size, hidden_size)
        self.bias_input_hidden = np.random.randn(1, hidden_size)
        self.weights_hidden_output = np.random.randn(hidden_size, output_size)
        self.bias_hidden_output = np.random.randn(1, output_size)

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def forward(self, X):
        self.hidden_output = self.sigmoid(np.dot(X, self.weights_input_hidden) + self.bias_input_hidden)
        self.final_output = self.sigmoid(np.dot(self.hidden_output, self.weights_hidden_output) + self.bias_hidden_output)
        return self.final_output

    def train(self, X, y, epochs, learning_rate):
        for epoch in range(epochs):
            output = self.forward(X)
            error = y - output
            loss = np.mean(np.square(error))
            d_output = error * output * (1 - output)
            error_hidden = d_output.dot(self.weights_hidden_output.T)
            d_hidden = error_hidden * self.hidden_output * (1 - self.hidden_output)
            self.weights_hidden_output += self.hidden_output.T.dot(d_output) * learning_rate
            self.bias_hidden_output += np.sum(d_output, axis=0, keepdims=True) * learning_rate
            self.weights_input_hidden += X.T.dot(d_hidden) * learning_rate
            self.bias_input_hidden += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate
            if epoch % 100 == 0:
                print(f"Epoch {epoch}, Loss: {loss}")

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Normalize features
X = StandardScaler().fit_transform(X)

# Split dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)

# Initialize neural network
input_size = X_train.shape[1]
hidden_size = 8
output_size = len(np.unique(y))
nn = NeuralNetwork(input_size, hidden_size, output_size)

# Train the neural network
nn.train(X_train, np.eye(output_size)[y_train], epochs=1000, learning_rate=0.1)

# Test the neural network
predictions = np.argmax(nn.forward(X_test), axis=1)
accuracy = np.mean(predictions == y_test)
print("Accuracy:", accuracy)
```

```python
#8..........................
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import adjusted_rand_score
from sklearn.preprocessing import StandardScaler

# Load Iris dataset
iris = load_iris()
X = iris.data
df = pd.DataFrame(X, columns=iris.feature_names)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(X_scaled, columns=iris.feature_names)

# Apply K-Means
kmeans = KMeans(n_clusters=3)
kmeans.fit(df_scaled)
df['KMeans Cluster'] = kmeans.labels_

# Apply EM algorithm
em = GaussianMixture(n_components=3)
em.fit(df_scaled)
df['EM Cluster'] = em.predict(df_scaled)

# Calculate Adjusted Rand Index to compare clustering results
ari = adjusted_rand_score(df['KMeans Cluster'], df['EM Cluster'])
print("Adjusted Rand Index (ARI) between K-Means and EM:", ari)

# Plotting
plt.figure(figsize=(12, 5))

# Plot K-Means clusters
plt.subplot(1, 2, 1)
plt.scatter(df['petal length (cm)'], df['petal width (cm)'], c=df['KMeans Cluster'], cmap='viridis', s=50, alpha=0.5)
#plt.scatter(kmeans.cluster_centers_[:, 2], r_cekmeans.clustenters_[:, 3], c='red', s=200, marker='X')
plt.title('K-Means Clustering')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# Plot EM clusters
plt.subplot(1, 2, 2)
plt.scatter(df['petal length (cm)'], df['petal width (cm)'], c=df['EM Cluster'], cmap='viridis', s=50, alpha=0.5)
#plt.scatter(em.means_[:, 2], em.means_[:, 3], c='red', s=200, marker='X')
plt.title('EM Clustering')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

plt.tight_layout()
plt.show()
```

```python
#10. Implement the non-parametric Locally Weighted Regression algorithm in order to fit data points.
#Select appropriate data set your experiment and draw graphs.
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(0)
X = np.linspace(0, 10, 100)
y = np.sin(X) + np.random.normal(0, 0.1, 100)

# Locally Weighted Regression function
def locally_weighted_regression(x_predict, X, y, tau):
    m = X.shape[0]
    weights = np.exp(-np.sum((X - x_predict) ** 2, axis=1) / (2 * tau ** 2))
    W = np.diag(weights)
    theta = np.linalg.inv(X.T @ W @ X) @ (X.T @ W @ y)
    prediction = x_predict @ theta
    return prediction

# Bandwidth parameter
tau = 0.5

# Prediction for the first 50 test points
predictions = [locally_weighted_regression(np.array([X[i]]), X.reshape(-1, 1), y, tau) for i in range(50)]

# Plot actual vs predicted values
plt.plot(X[:50], y[:50], label='Actual')
plt.plot(X[:50], predictions, label='Predicted')
plt.title('Locally Weighted Regression on Synthetic Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
